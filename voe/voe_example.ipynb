Do vibration of effects: 

```python
voe = {'beta': [], '-np.log(p)': [], 'num_predictors': []}
for k in np.arange(1,X.shape[1]):
    for comb in combinations(np.arange(2,X.shape[1]), k):
        model = sm.OLS(y, X[:,[0]+list(comb)])
        results = model.fit()
        voe['beta'].append(results.params[0])
        voe['-np.log(p)'].append(-np.log10(results.pvalues[0]))
        voe['num_predictors'].append(k)
voe = pd.DataFrame(voe)
```

Now plot

```python
sns.kdeplot(data =voe["beta"], data2=voe["-np.log(p)"], shade=False, alpha=0.5)
sns.scatterplot(x =voe["beta"], y = voe["-np.log(p)"], hue=voe['num_predictors'], 
                palette='coolwarm_r')
plt.gca().set_title('VOE of ' +var_name)
```
